```python
query = """
SELECT *
FROM crime_scene_report
WHERE city = "SQL City"
ORDER BY date;
"""

pd.set_option('display.max_colwidth',None)   #set the width for data to be displated set it to maximum

pd.read_sql_query(query,conn)
```

Security footage shows that there were 2 witnesses. The first witness lives at the last house on "Northwestern Dr.". The second witness, named Annabel, lives somewhere on "Franklin Ave".

So as per the report there are 2 witnesses. Let's look for them.

```python
#checking persnal details
query_2 = '''
SELECT *
FROM person
WHERE address_street_name = "Northwestern Dr"
ORDER BY address_number desc;
'''
pd.read_sql_query(query_2, conn)[:1]
```

```python
query_3 = '''
SELECT *
FROM person
WHERE name like '%Annabel%' AND address_street_name = "Franklin Ave";
'''
pd.read_sql_query(query_3, conn)
```

```python
#Viewing the interview of al the witnesses taken after the murder.
query_4 = '''
SELECT *
FROM interview
WHERE person_id = 14887 OR person_id = 16371;
'''
pd.read_sql_query(query_4, conn)
```

So we got 2 clues here:-

Killer is a man & a member of the gym with a status of gold & having a membership number starting with "48Z" and left in a car with a number plate of "H42W". He was working out in the gym on the 9th of Jan

```python
#Checking the gym database with above details
query_5 = '''
SELECT *
FROM get_fit_now_check_in
WHERE membership_id like "%48Z%" AND check_in_date = 20180109
order by check_in_date;
'''
pd.read_sql_query(query_5, conn)
```

```python
#Let's check the car details by the above details
query_6 = '''
SELECT *
FROM drivers_license
WHERE plate_number like "%H42W%"
'''
pd.read_sql_query(query_6, conn)
```

Got 2 males with plate number containing H24W

```python
#checking personal details of both the males from the query
query_7 = '''
SELECT *
FROM person
WHERE license_id = "423327" OR license_id = "664760";
'''
pd.read_sql_query(query_7, conn)
```

```python
#hecking which of this two are a member of the gym?
query_8 = '''
SELECT *
FROM get_fit_now_member
WHERE person_id = "51739" OR person_id = "67318";
'''
pd.read_sql_query(query_8, conn)
```
