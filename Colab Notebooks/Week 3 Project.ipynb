# **FIXED COLUMN NAMES**

```python
#fixed column names
data1.columns = data1.columns.str.lower().str.replace(' ', '_')   #data1.columns.str.lower() converts names of the column in lower case, using str.replace() also permits to replace the space in "easy apply to be replaced with an underscore"
print(data1.columns)
```

# **AGE COLUMN**

```python
#Age Column
data1['age']  #to compare clean Age column
```

```python
average_age = data1.age.mean()
print (average_age)
```

```python
#Fix age Column
data1['age'] = data1['age'].fillna(average_age)   #fillna() functions allows us to fill the null values with the average value & update the original dataset with these values
data1['age'] = data1['age'].round(decimals=1)     #rounding off the values
data1['age']
```

```python
data1['age'].isna().sum()
```

# **SALARY COLUMN**

```python
data1['salary'] = data1['salary'].apply(lambda x:x.replace('$', ''))
data1['salary'] = data1['salary'].apply(lambda x:x.replace('k', ''))
print(data1['salary'])
```

```python
data1.head()
```

**Changed data type of the age column to 'int'**

```python
data1['age'] = data1['age'].astype(int)
print(data1['age'])
```

# **ESTABILISHED COLUMN**

```python
#Clean Estabilished Column
data1['established'] = data1['established'].replace(-1,np.nan)
data1['established'].fillna('Unknown',inplace=True)
print(data1['established'])
```

# **LOCATION COLUMN**

```python
#Location Column
data1['location'] = data1.location.astype
```

```python
#Cleaning the location column and splitting the city from the city short name

data1['location'] = data1['location'].str.replace("Australia Aus", "Australia, Aus")
data1['location_city'] = data1['location'].str.split(",", expand=True).get(0)
data1['city_sign'] = data1['location'].str.split(",", expand=True).get(1)
data1.drop('location', inplace=True, axis=1)
```

# **RATING COLUMN**

```python
#Rating Column
#Cleaning the RAting Column and replacing special characters with Nan Values
data1['rating'] = data1["rating"].replace(-1,np.nan)

#Now replace those nan values with the ratind mean()
data1['rating'] = data1["rating"].fillna(data1['rating'].mean()).round(decimals=1)

print(data1['rating'])
```

# **EASY APPLY COLUMN**

```python
#Cleaning the easy_apply
data1['easy_apply'] = data1['easy_apply'].replace("-1", "True")
print(data1['easy_apply'])
```

# **TURN 'item_price' INTO FLOAT**

```python
#Turn the item_price into float
data2['item_price'] = data2['item_price'].replace('\D', '', regex=True)
data2['item_price'] = data2['item_price'].replace('', np.nan)  #Replace empty strings
data2['item_price'] = data2['item_price'].astype(float)
print(data2['item_price'])
```

# **Quantity of each item**

```python
#Quantity of Each item
item_quantities = data2['item_name'].value_counts().reset_index()
item_quantities.columns = ['item_name', 'quantity']
print(item_quantities)
```

# **What was the most ordered item in the choice_description column?**

```python
#What was the most ordered item in choice descrption column
choice_description_column = data2['choice_description']

if not choice_description_column.dropna().empty:
  most_ordered_item = choice_description_column.dropna().value_counts().idxmax()
  print("The most ordered item is : ", most_ordered_item)

else:
  print("No null-values in the 'choice_description column.")


```

# **How many items were ordered in total?**

```python
#How many items were ordered in total
total_ordered_items = data2['quantity'].sum()

print("Total number of items ordered:", total_ordered_items)
```

# **How much was the revenue for the period in the dataset?**

```python
#How much was the revenue for the period in the data set?
# Clean the 'item_price' column and convert it to a numeric type
data2['item_price'] = data2['item_price'].replace('[\$,]', '', regex=True).astype(float)

# Calculate the revenue by multiplying quantity and item price, then summing the results
total_revenue = (data2['quantity'] * data2['item_price']).sum()

print("Total revenue for the period:", total_revenue)
```

# **How many orders were made in the period?**

```python
#How many orders were made in the period?
total_orders = data2['order_id'].nunique()

print("Total number of orders made in the period:", total_orders)
```

# **What is the average revenue amount per order?**

```python
# Calculate the average revenue per order
average_revenue_per_order = round(total_revenue / total_orders, 2)

print("Average revenue amount per order:", average_revenue_per_order)
```
